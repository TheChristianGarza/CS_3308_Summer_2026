```python
#Esmeralda Nuno Cortez and Christian Garza
#June 28th, 2026

import numpy as np
import math

def eval_expr(expr):
  dictionary = {
      "pi": math.pi,
      "e": math.e,
      "sqrt": math.sqrt
  }
  return eval(expr, {"__builtins__": None}, dictionary)

def gaussian_elimination(mat):
    #Converting the matrix into a numpy array that holds float variables will make the numbers be precise
    mat = np.array(mat, dtype=float)
    n = len(mat)

    #Forward elimination with partial pivoting
    for i in range(n):
        #Finds the row wiht the largest absolute value in the column it is on
        max_row = i + np.argmax(np.abs(mat[i:n, i]))

        #Swaps the current row with the maximum row
        if max_row != i:
            mat[[i, max_row]] = mat[[max_row, i]]

        #Checking to see if a diagonal has no solution by if a diagonal is zero or nearing zero
        if abs(mat[i, i]) < 1e-9:
            print("Error: The system has no unique solution.")
            return None

        #Makes sure the numbers under the pivot are zero
        for k in range(i + 1, n):
            factor = mat[k, i] / mat[i, i]
            mat[k, i:] -= factor * mat[i, i:]

    #Back substitution
    solution = np.zeros(n) #Creates an array with all zeros
    for i in range(n - 1, -1, -1):
        #Start with the constant term
        solution[i] = mat[i, n] - np.dot(mat[i, i+1:n], solution[i+1:n])
        #Divide by the pivot coefficient
        solution[i] /= mat[i, i]

    return solution
#Just to make the Equation output look good
def align_equations(all_expr_lists, all_const_exprs):
    raw_terms = [
        [f"{expr}*x{j+1}" for j, expr in enumerate(expr_list)]
        for expr_list in all_expr_lists
    ]
#Compute max width PER COLUMN across ALL equations
    num_vars = len(raw_terms[0])
    col_widths = [
        max(len(raw_terms[row][col]) for row in range(len(raw_terms)))
        for col in range(num_vars)
    ]
#Build aligned equations
    aligned_equations = []
    for row_idx, row_terms in enumerate(raw_terms):
        padded_terms = [
            row_terms[col].ljust(col_widths[col])
            for col in range(num_vars)
        ]
        left_side = " + ".join(padded_terms)
        aligned_equations.append(f"{left_side} = {all_const_exprs[row_idx]}")
    return aligned_equations


#Just to make the solution output look good
def build_solution_string(solution):
    sol_strings = [f"{val:.6f}" for val in solution]

  #Find longest solution value for alignment
    max_len = max(len(s) for s in sol_strings)
    lines = []
    for i, s in enumerate(sol_strings):
        padded = s.rjust(max_len)
        lines.append(f"x{i+1} = {padded}")
    return lines

#Test carry over to printing

def main():
  n = int(input("Enter the number of equations: "))
  mat = []

  equations_expr_lists = []
  const_exprs = []

  print("Coefficients: ")
  print("For an equation like: a1*x1 + a2*x2 + ... = Constant\n")

  for i in range(n):
      print(f"Equation {i + 1}:")
      coeff_exprs = []
      coeff_values = []

      for j in range(n):
          expr = input(f"   Coefficient for x{j + 1}: ")
          coeff_exprs.append(expr)
          coeff_values.append(eval_expr(expr))

      const_expr = input("   Constant term: ")
      const_value = eval_expr(const_expr)
      mat.append(coeff_values + [const_expr])

      equations_expr_lists.append(coeff_exprs)
      const_exprs.append(const_expr)

      print()

  aligned_equations = align_equations(equations_expr_lists, const_exprs)

  solution = gaussian_elimination(mat)

  if solution is not None:
    print()
    print("Full Linear System: ")
    for i, eq in enumerate(aligned_equations):
      print(f"Eq {i+1}: {eq}")

    print("\nSolution: ")
    aligned_solution = build_solution_string(solution)
    for line in aligned_solution:
      print(line)

if __name__ == "__main__":
    main()
```
